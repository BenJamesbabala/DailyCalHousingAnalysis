```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.dates as mdates
import csv
%matplotlib inline
```

```python
avg_ppr = []
addresses = []

with open("rent_collection_board_avg_ppr.csv") as f:
    rentreader = csv.DictReader(f)
    for row in rentreader:
        avg_ppr.append(float(row['avg_ppr']))
        addresses.append(row['address'])
        
```

```python
#Histogram of average prices
plt.figure(figsize=(12,6), dpi=80)
plt.hist(avg_ppr, bins=100)
plt.xlabel("Average Price Per Room in Berkeley")
plt.ylabel("Frequency of Price")
plt.show()
```

```python
addresses_2 = []
price = []
movein_t = []
movein_d = []
bedrooms = []

with open("rent_collection_board.csv") as f:
    rentreader = csv.DictReader(f)
    for row in rentreader:
        addresses_2.append(row['address'])
        price.append(float(row['price']))
        movein_t.append(float(row['movein_t']))
        movein_d.append(row['movein_d'])
        bedrooms.append(float(row['bedrooms']))
        
```

```python
mpl_data = mdates.epoch2num(movein_t)

# plot it
fig, ax = plt.subplots(1,1)
fig.set_size_inches(12, 6, forward=False)
ax.hist(mpl_data, bins=50, color='lightblue')
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m.%d.%y'))
plt.show()
```
