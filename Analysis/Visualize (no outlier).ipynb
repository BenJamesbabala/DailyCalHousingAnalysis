```python
plt.scatter(df['sqft per room'], df['price per room'])
plt.xlabel('Room size')
plt.ylabel('price per room')
```

```python
curr = df['sqft per room'].values[np.newaxis].T
ones = np.array([1]*len(curr))[np.newaxis].T

use = np.hstack((curr, ones))
labels = df['price per room'].values.T
coeffs = np.linalg.lstsq(use, labels)[0]

plt.scatter(df['sqft per room'], df['price per room'])
plt.plot(df['sqft per room'], coeffs[0]*df['sqft per room'] + coeffs[1])
plt.xlabel('Sqft per room')
plt.ylabel('price per room')
plt.title("Least Squares prediction")
```
